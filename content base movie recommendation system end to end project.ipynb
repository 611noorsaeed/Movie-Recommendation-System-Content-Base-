```python
import numpy as np
import pandas as pd
```

```python
movie = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movie.head()
```

```python
credits.head()
```

# merging both dataset

```python
movies = movie.merge(credits, on='title')
```

```python
movies.head()
```

# keeping important columns

```python
movies.columns
```

```python
movies = movies[['id','title','overview','keywords','genres','cast','crew']]
```

```python
movies
```

# checking null vals

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

# working with overview `

```python
movies.iloc[0].overview
```

```python
# it is a string so convert is into list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['overview'][0]
```

# working with keywords

```python
movies['keywords'][0]
```

```python
import ast  # use to convert string to integer
def keywords(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['keywords'] = movies['keywords'].apply(keywords)
```

```python
movies['keywords']
```

# working with genres

```python
movies['genres'][0]
```

```python
import ast  # use to convert string to integer
def genres(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(genres)
```

```python
movies['genres']
```

# working with cast

```python
movies['cast'][0]
```

```python
import ast  # use to convert string to integer
def cast(obj):
    l = []
    # intersted in top three cast
    count = 0
    for i in ast.literal_eval(obj):
        if count != 3:
            l.append(i['name'])
            count+=1
        else:
            break
    return l

movies['cast'] = movies['cast'].apply(cast)
```

```python
movies['cast']
```

# working with crew

```python
movies['crew'][0]
```

```python
import ast  # use to convert string to integer
def crew(obj):
    l = []
    # intersted in top three cast
    count = 0
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew'] = movies['crew'].apply(crew)
```

```python
movies['crew']
```

# concatinating last fours cols into one`

```python
movies['tags'] = movies['overview'] + movies['cast'] + movies['crew'] + movies['keywords']
```

```python
movies =  movies[['id','title','tags']]
```

```python
movies
```

# removing spaces form tags

```python
movies['tags'] = movies['tags'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies['tags'][0]
```

# applying stemming

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
movies['tags']
```

```python
def stemming(text):
    l = []
    for i in text:
        l.append(ps.stem(i))
    return " ".join(l)
```

```python
movies['tags'] = movies['tags'].apply(stemming)
```

```python
movies['tags'][10]
```

# Vectorization code

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(max_features=500, stop_words='english')
```

```python
vectors = vectorizer.fit_transform(movies['tags']).toarray()
```

```python
vectors
```

```python
vectorizer.get_feature_names()
```

# Calculating distances

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
movies[movies['title']=="Avatar"]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
```

```python
movies.iloc[100].title
```

```python
def Recommendation_system(movie):
    movie_index = movies[movies['title']== movie].index[0]
    distances = sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
    
    for i in distances[1:20]:
        print(movies.iloc[i[0]].title)
```

```python
Recommendation_system('Avatar')
```
